## Exercício 1

Escreva um programa em Python que receba a nota de um aluno (em escala de 0 a 100) e informe a sua classificação conforme a tabela abaixo:
- Nota de 0 a 49: Reprovado
- Nota de 50 a 69: Recuperação
- Nota de 70 a 100: Aprovado

Exemplo de uso:

```
Digite o valor da nota (de 0 a 100): 55
Status do aluno: recuperação.
```

**Entrada:** Nota do aluno (valor ``int``).

**Saída:** Mensagem de status do aluno.

````python
def exercicio1(nota):
  nota = int(input('Digite o valor da nota (de 0 a 100): '))
  if nota >= 0 and nota <= 49:
    return f'Status do aluno: reprovado'
  elif nota >= 50 and nota <= 69:
    return f'Status do aluno: recuperação'
  elif nota >= 70 and nota <= 100:
    return f'Status do aluno: aprovado'
  elif nota > 100:
    return f'Nota incorreta por ultrapassar 100, uma nota é valida se estiver entre 0 e 100'

exercicio1(0)
````

O seu programa deve calcular e imprimir uma sequência de números reais $B_1, B_2, \ldots B_{n}$ em que $B_1 = 0$ e $B_i = A_i\cdot A_{i-1}$ para $i \geq 2$. Isto é, $B_2 = A_1\cdot A_{2}$; $B_3 = A_2\cdot A_{3}$, e assim por diante.


Note que para resolver o problema, você _pode_ guardar todos os valores $A_i$ em uma lista. Porém, também é possível solucionar o problema sem usar listas.
