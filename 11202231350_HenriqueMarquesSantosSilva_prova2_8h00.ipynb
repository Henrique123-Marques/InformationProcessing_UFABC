Neste exercício, você vai criar uma função para contar, em uma lista de números, quantos elementos são maiores que a média.

Em primero lugar, crie uma função ``media(seq)`` que recebe uma lista de números ``seq`` como entrada e **devolve o valor da média** dos números **como um número em ponto flutuante**.

Você deve usar apenas operações aritméticas básicas (+, -, /, *) e as estruturas estudadas na disciplina para implementar a função. Não é permitido usar funções externas, como ``sum``.



```python
# Use esta célula para implementar a sua função de média.
def media(seq):
  soma = 0
  tamanho_lista = len(seq)

  #Soma dos indices da lista
  for i in seq:
     soma += i
  #Media da lista
  media_seq = soma/tamanho_lista
  return media_seq
```

## Exercício 2

Em certo _videogame_, um personagem pode se mover sobre um cenário em duas dimensões. Nós representamos a posição do personagem por um par de coordenadas $(x, y)$ em um plano cartesiano. As coordenadas são armazenadas em uma lista de tamanho 2 (o primeiro índice da lista corresponde ao eixo $x$ e o segundo ao eixo $y$).

O personagem pode se movimentar para a esquerda (E), direita (D), frente (F) ou trás (T).

- A cada movimento para a **direita**, a coordenada $x$ do personagem é **aumentada de 1 metro**.
- A cada movimento para a **esquerda**, a coordenada $x$ do personagem é **diminuída de 1 metro**.
- A cada movimento para **frente**, a coordenada $y$ do personagem é **aumentada de 1 metro**.
- A cada movimento para **trás**, a coordenada $y$ do personagem é **diminuída de 1 metro**.

Uma sequência de movimentos é representada por um _string_ contendo somente as letras "E", "D", "F" ou "T".

Crie uma função ``deslocamento(movimentos)`` que recebe um _string_ ``movimentos`` que representa uma sequência de movimentos do personagem; este _string_ é formado apenas pelas letras "E", "D", "F" e "T". Sua função deve devolver uma lista de tamanho 2, representado o correspondente _deslocamento_ do personagem em termos de coordenadas $(x, y)$.

Exemplos de uso:
- ``deslocamento("EEF")`` deve devolver a lista ``[-2, 1]``.
- ``deslocamento("DTD")`` deve devolver a lista ``[2, -1]``.
- ``deslocamento("DDTTEEFF")`` deve devolver a lista ``[0, 0]``.




```python
# Implemente a sua função usando esta célula.

def deslocamento(movimentos):
  #Coordenadas
  x, y = 0, 0

  #Analise dos movimentos
  for movimento in movimentos:
      if movimento == 'E':
          x -= 1
      elif movimento == 'D':
          x += 1
      elif movimento == 'F':
          y += 1
      elif movimento == 'T':
          y -= 1
  posicao = [x, y]
  return posicao
```

```python
#Testes
if deslocamento(('EEF')) != [-2, 1]:
  print('Erro')
if deslocamento(('DTD')) != [2, -1]:
  print('Erro')
if deslocamento(('DDTTEEFF')) != [0, 0]:
  print('Erro')
```

```python
#Resultado final, valor recebido na funcao
deslocamento(('EEF'))
```
