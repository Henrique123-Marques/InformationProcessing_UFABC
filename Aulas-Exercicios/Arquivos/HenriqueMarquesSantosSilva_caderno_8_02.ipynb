## Exercício 5

Altere o programa do exercício anterior, para que o usuário possa escolher um arquivo de saída onde armazenar os nomes dos estados. Inicialmente, seu programa deve perguntar ao usuário qual o nome do arquivo de saída desejado. O programa precisa abrir este arquivo para **escrita** e então escrever os nomes do estados, um em cada linha.

Lembre-se de fechar os arquivos de entrada e de saída antes de finalizar seu programa.

````python
def arquivo_saida():
  with open('cidades.txt', 'r') as c:
    lista_estados = c.readlines()
    lista_estados = [estado.strip() for estado in lista_estados[0:]]

  pergunta = str(input('Qual o nome do arquivo de saída?: Exemplo: arquivo.txt '))
  with open(pergunta, 'w') as f:

    for estado in lista_estados:
      estado = estado.split(',')
      f.write(estado[1] + '\n')
      print(estado[1])


arquivo_saida()
````

## Exercício 6

Usando o mesmo arquivo *cidades.txt*, faça um programa que imprime apenas os nomes de cidade *de um determinado estado*. Primeiramente, seu programa deve solicitar ao usuário que digite um nome de estado. Seu programa deve filtrar as linhas, imprimindo apenas o nome das cidades localizadas no estado escolhido pelo usuário. Ao final do programa, deve ser mostrado o número de cidades encontradas. Por exemplo, caso o usuário selecione *Rio de Janeiro* como estado, seu programa deve ter uma saída como:

```
Rio de Janeiro
São Gonçalo
Duque de Caxias
3 cidades encontradas
```

````python
def escolher_estados():
  #Lista de estados sendo mostrada
  with open('cidades.txt', 'r') as c:
    estados = c.readlines()
    estados = [estado.strip() for estado in estados[0:]]
    novos_estados = list()
    print('LISTA DE ESTADOS:')
    for estado in estados:
      estado = estado.split(',')
      #print(f'{estado[1]}')
      novos_estados.append(estado[1])

#Mensagem para se o estado esta na lista
  cidades_do_estado = list()
  msg = str(input('Digite o nome de um estado: '))
  if msg in novos_estados:
    print('\n')
    for estado in estados:
      estado = estado.split(',')
      if estado[1] == msg:
        cidades_do_estado.append(estado[0])
    print(f'Estado: {msg}; Suas cidades: {cidades_do_estado}\n')
    print(f'{len(cidades_do_estado)} cidades encontradas')
  else:
    print(f'\nEstado nao encontrado')

#Listagem das cidades nesse estado encontrado:

escolher_estados()
````
