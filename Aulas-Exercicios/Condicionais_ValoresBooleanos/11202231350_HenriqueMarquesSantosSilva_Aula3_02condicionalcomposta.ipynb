### Exemplo 1

Escreva um programa que recebe os coeficientes de uma equação quadrática $ax^2 + bx + c = 0$ e calcula as raízes da equação. O programa deve imprimir uma mensagem de erro caso o discriminante seja negativo e se comportar corretamente caso no $a = 0$.

````python
# vamos importar o módulo math para usar a função sqrt (raiz quadrada)
import math

# coeficientes da equação
a = 0
b = 4
c = 1

# cálculo do discriminante
dd = b*b - 4*a*c

# cálculo das raízes
if a == 0:
    raiz = -c/b
    print("Equação de primeiro grau. x =", raiz)
elif dd < 0:
    print("Discriminante negativo. Não é possível encontrar raízes reais")
else:
    raiz1 = (-b + math.sqrt(dd))/(2*a)
    raiz2 = (-b - math.sqrt(dd))/(2*a)
    print("Primeira raiz. x =", raiz1)
    print("Segunda raiz. x =", raiz2)



````
