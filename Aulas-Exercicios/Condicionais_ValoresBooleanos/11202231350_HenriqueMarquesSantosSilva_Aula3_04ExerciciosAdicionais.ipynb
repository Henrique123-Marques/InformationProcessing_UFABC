# Exercício 2

Em uma certo um país, o imposto de renda devido é calculado de maneira progressiva, de acordo com a regra abaixo:
- até 500 dinheiros, a alíquota é zero (isenção).
- acima de 500 até 1000 dinheiros, o imposto é calculado como:
Imposto = (Renda - 500)*0.1
- acima de 1000 dinheiros, o imposto é calculado como:
Imposto = 50 + (Renda - 1000)*0.2.

A alíquota efetiva do imposto é dada pela razão (Imposto)/(Renda). Faça um programa que recebe um valor de renda digitado pelo usuário (expresso por um número em ponto flutuante) e imprime o valor do imposto devido e da alíquota efetiva.

```python
# resolva o exercício 2 nesta célula
renda = float(input('Digite o valor da renda: '))
if renda <= 500:
  print('Isento do imposto de renda, por ter uma renda menor que 500 dinheiros')
elif renda > 500 and renda < 1000:
  imposto = (renda-500)*0.1
  print(f'O valor da renda está entre 500 e 1000 dinheiros e vale {renda}, com isso o imposto de renda é: {imposto}. Logo a alíquota vale: {imposto/renda:.2f} com duas casas decimais')
elif renda>1000:
  imposto = 50+(renda-1000)*0.2
  print(f'O valor da renda passa de 1000 dinheiros, sendo ela {renda}, com isso o imposto de renda é: {imposto}. Logo a alíquota vale: {imposto/renda:.2f} com duas casas decimais')
```
