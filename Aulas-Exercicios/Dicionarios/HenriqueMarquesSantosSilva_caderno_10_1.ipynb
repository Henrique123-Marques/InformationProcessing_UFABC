## Exercício 1

Escreva uma função `conta_inteiros` que recebe como argumento uma lista `t` de números inteiros e devolve um dicionário `d` que relaciona cada número inteiro distinto em `t` e seu número de ocorrências. Isto é, as chaves de `d` são elementos de `t`, e o valor associado a cada chave é a quantidade de vezes que a chave aparece em `t`. Caso `t`seja uma lista vazia, sua função deve devolver um dicionário vazio também (`d = {}`).


````python
def conta_inteiros(t):
    contagem = {}
    for numero in t:
      if numero in contagem:
        contagem[numero] += 1
      else:
        contagem[numero] = 1
    return contagem

# se sua função estiver correta, ela deve passar
# pelos testes abaixo
if conta_inteiros([]) != {}:
    print("Erro: lista vazia deve gerar dicionário vazio")
else:
    print("Teste 1: ok!")
entrada = [4, 6, 4, 5, 2, 3, 2]
esperado = {2:2, 3:1, 4:2, 5:1, 6:1}
obtido = conta_inteiros(entrada)
if obtido != esperado:
    print("Erro!")
    print("Esperava: ", esperado)
    print("Obteve: ", obtido)
else:
    print("Teste 2: ok!")

````
