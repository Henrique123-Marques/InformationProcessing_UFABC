Em um exercício anterior, você criou um programa para o jogo _Pedra, Papel e Tesoura_. Crie uma função ``vencedor(jogador1, jogador2)`` que recebe a opção feita pelo jogador 1, a opção feita pelo jogador 2 e que calcula o vencedor, devolvendo:
- 0 (zero) caso haja empate

- 2 (dois) caso o vencedor seja o segundo jogador.

Crie uma nova versão do seu programa do jogo, agora usando a função ``vencedor`` para decidir o resultado.

## Exercício 2

Crie uma função ``centraliza(texto, num_colunas)`` que recebe um _string_ texto e um número inteiro positivo ``num_colunas`` e produz uma versão centralizada do texto, usando espaços, considerando uma largura total de ``num_colunas``. Por exemplo, ``centraliza("UFABC", 11)`` deve produzir o _string_ ``"   UFABC   "`` (isto é, 3 espaços, seguidos por ``"UFABC"``, seguidos por mais 3 espaços, totalizando 11 caracteres).
Lembre-se que:
- a função ``len(meu_string)`` devolve o número de caracteres do argumento ``meu_string``.
- a operação ``meu_string*n`` produz um novo _string_ composto por n repetições de ``meu_string``.

```python
def centraliza(texto, num_colunas):
  return ((' '*num_colunas) + texto + (' '*num_colunas))

centraliza('HMSS', 4)
#centraliza(str(input('Digite uma palavra ou letra: ')), int(input('Digite um valor para o número de colunas: ')))
```

## Exercício 3

Crie uma função ``combina_string(a, b)`` que recebe dois _strings_ ``a`` e ``b`` e devolve a concatenação _menor+maior+menor_ em que _menor_ é o _string_ mais curto entre ``a`` e ``b`` e _maior_ é o _string_ mais longo. Por exemplo:
- ``combina_string("UFABC", "~")`` devolve ``~UFABC~``.
- ``combina_string("+++", "HOJE")`` devolve ``+++HOJE+++``.

```python
def combina_string(a,b):
  if a>b:
    return b + a + b
  else:
    return a + b + a

combina_string('HMSS', '&')
```

## Exercício 4

Crie uma função ``preenche_esq(num, largura)`` que recebe dois número inteiros positivos ``num`` e ``largura``. A função deve devolver uma representação como _string_ de ``num`` preenchida com zeros à esquerda até atingir a largura determinada. Caso o número de dígitos de ``num`` seja maior que ``largura``, a função deve devolver apenas ``num`` convertido para _string_, sem preenchimento. Por exemplo:
- ``preenche_esq(10, 4)`` devolve "0010".
- ``preenche_esq(123, 3)`` devolve "123".
- ``preenche_esq(1035, 3)`` devolve "1035".

Pense em primeiro converter ``num`` para _string_. Você pode usar a função ``str`` para isso, isto é, ``str(num)``. Em seguida, calcule o tamanho da representação de ``num`` como _string_. Com isso, você pode determinar quantos zeros devem ser adicionados.

```python
def preenche_esq(num, largura):

  num_string = str(num)
  digitos = len(num_string)

  if digitos >= largura:
    return num_string
  else:
    zeros = largura - digitos
    preenchimento = '0' * zeros + num_string

  return preenchimento


preenche_esq(10,5)
```
