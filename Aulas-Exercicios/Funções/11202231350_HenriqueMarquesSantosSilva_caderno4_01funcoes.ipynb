## Exercício 5

O cálculo de *juros compostos* envolve um valor monetário inicial, denominado *capital*, uma taxa de juros *j* por período e o tempo total *t* transcorrido. Ao final desse tempo, o valor corrigido pelos juros, denominado *montante* é determinado pela fórmula:

$$\text{montante} =  \text{capital}(1 + j)^{t}$$

Já no caso de *juros simples*, o valor da taxa de juros é aplicada sempre ao valor do capital inicial, resultando na fórmula:

$$\text{montante} =  \text{capital}(1 + jt)$$

1. Crie uma funçao ``juro_composto`` que recebe um valor inicial monetário, um valor de taxa (mensal) de juros e uma quantidade de meses e devolve o valor do montante final. Use a fórmula de juros compostos.
2. Use a função acima para calcular o montante final para um empréstimo de R\$ 1000,00, com taxa de 0.5\% ao mês e tempo total de 12 meses.
3. Calcule a taxa de juros simples que resultaria no mesmo montante calculado no item anterior, nas mesmas condições (valor  inicial de mil reais,  tempo total de 12 meses). Você deve resolver esta questão usando código Python.

Ao resolver a questão, reflita sobre o seguinte ponto: valores monetários não se comportam exatamente como números reais. Por exemplo, não é possível ir ao banco e sacar R\$ 1,3958672; os valores monetários usuais são limitados a duas casas decimais. Como você acha que o seu programa deveria lidar com essa questão? Se for necessário "arredondar" valores dentro do seu programa, quando isso deveria ser feito?

````python
#1. Cálculo do montante pela formula de juros
def juro_composto(capital, j, t):
  montante = (capital*(1+j)**t)
  return f'1. O valor monetário inicial (capital) digitado foi R${capital}, a taxa mensal digitada foi {j}%, e o tempo mensal digitado foi {t} meses, então o montante vale: R$ {montante}'

juro_composto(float(input('Digite um valor inicial monetário (capital): ')), float(input('Digite um valor taxa mensal (j): ')), float(input('Digite a quantidade de meses (t): ')))
````

````python
#2. Cálculo do montante final para um emprestimo de 1000, com taxa de 0.5% ao mes e tempo de 12 meses
def juro_composto2(emprestimo, taxa_mensal, tempo):
  montante_final = (emprestimo*(1+taxa_mensal)**tempo)
  return f'2. O valor do empréstimo digitado foi R${emprestimo}, a taxa mensal digitada foi {taxa_mensal}%, e o tempo mensal digitado foi {tempo} meses, então o montante vale: R$ {montante_final:.2f}'

juro_composto2(1000, 0.5, 12)
````

````python
#3. Cálculo da taxa de juros simples para um emprestimo de 1000, com taxa de 0.5% ao mes e tempo de 12 meses
def taxas_juros_simples(emprestimo, taxa_mensal, tempo):
  taxa_simples = (emprestimo*(1+taxa_mensal*tempo))
  return f'2. O valor do empréstimo digitado foi R${emprestimo}, a taxa mensal digitada foi {taxa_mensal}%, e o tempo mensal digitado foi {tempo} meses, então o montante vale: R$ {taxa_simples:.2f}'

taxas_juros_simples(1000, 0.5, 12)
````
