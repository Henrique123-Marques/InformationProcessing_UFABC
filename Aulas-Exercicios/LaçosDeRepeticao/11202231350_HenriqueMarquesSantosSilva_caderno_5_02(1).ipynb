
## Exercício 1

A média aritmética de um conjunto $\{x_1, x_2, \ldots, x_N\}$ de $N$ números reais é dada por:

$\text{média}(x_1, x_2, \ldots, x_N) = \frac{1}{N}\sum_{i=1}^{N} x_i$

A _variância_ $\sigma^2$ de um conjunto $\{x_1, x_2, \ldots, x_N\}$ é uma medida da _variabilidade_ do conjunto de números e pode ser calculada como:

$\sigma^2(x_i, x_2,\ldots,x_N) = \frac{1}{N}\sum_{i=1}^{N} (x_i^2) - \big(\text{média}(x_i, x_2,\ldots,x_N)\big)^2$

Faça um programa que recebe 5 números reais digitados pelo usuário e que imprime o valor da média dos números e o valor da variância. Como passos intermediários para resolver esse problema, seu programa deve:

- calcular a soma dos números ($\sum_i x_i$)
- calcular a soma dos quadrados dos números ($\sum_i x_i^2$)

```python
#Bibliotecas usadas
import math
import numpy as np

lista = list() #Lista para armazenar os 5 valores
def exercicio1(n, soma, quadrados, media, varianca):
  for i in range(5):
    n = int(input('Digite um numero: '))
    lista.append(n)
    soma = sum(lista)
    quadrados = sum([x**2 for x in lista])
    media = soma/5
    varianca = np.var(lista) #Metodo np.var calcula a varianca dos valores
  return f'Com base nos numeros digitados: Soma: {soma}. Soma dos quadrados: {quadrados}; Média: {media}; Variância: {varianca}'


exercicio1(0,0,0,0,0)
```

## Exercício 2

A média _harmônica_ $\overline{h}$ de um conjunto $\{x_1, x_2, \ldots, x_N\}$ de $N$ números reais é dada por:

$\overline{h}(x_1, x_2, \ldots, x_N) = N \cdot \big(\sum_{i=1}^{N} \frac{1}{x_i}\big)^{-1}$

Faça um programa que recebe 5 números reais digitados pelo usuário e que imprime o valor da média _harmônica_ dos números. Como passos intermediários para resolver esse problema, seu programa deve:

- calcular o inverso de cada número ($x_i^{-1}$)

- calcular a _soma_ dos inversos ($\sum_i x_i^{-1}$)


```python
import math

lista = list()
def exercicio2(n, inverso, media_harmonica):
  for i in range(5):
    n = int(input('Digite um numero: '))
    lista.append(n)
    inverso = [1/x for x in lista]
    media_harmonica = 5/sum(inverso)

  return f'Inverso dos valores digitados: {inverso}, soma dos inversos: {sum(inverso)}, Média harmonica: {media_harmonica}'

exercicio2(0,0,0)
```
