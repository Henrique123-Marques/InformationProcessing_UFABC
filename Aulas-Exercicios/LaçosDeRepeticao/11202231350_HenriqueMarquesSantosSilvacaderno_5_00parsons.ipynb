## Problema 1

A célula abaixo contém uma função ``conta_zeros`` que recebe uma lista de números inteiros e devolve a quantidade de zeros que existem na lista. A célula seguinte contém alguns exemplos de utilização da função.

```python
def conta_zeros(lista):
  #Incializacao das variaveis
  indice = 0
  contador = 0
  num_itens = len(lista)

  #Enquanto o indice da lista for menor que o numero de items adicioone um indice
  while indice < num_itens:
    #Se o indice presente na lista de numeros for zero adicione cada um contador
    if lista[indice] == 0:
      contador = contador + 1
    indice = indice + 1
  return contador #retorne a quantidade de zeros

print(conta_zeros([]))
print(conta_zeros([0, 1, 2, 3]))
print(conta_zeros([3, 2, 1, 0]))
print(conta_zeros([2, 1, 0, 4, 3, 0, 5]))
print(conta_zeros([1, 2, 3]))
```

## Problema 2

A célula abaixo contém uma função ``busca`` que recebe uma lista de itens e um valor ``elem``. A função busca o valor dentro da lista; caso o valor seja encontrado, a função devolve o _índice_ do valor. Caso existam múltiplas ocorrências do valor, a função devolve o índice da primeira ocorrência. Caso o valor não seja encontrado, a função devole ``None``. A célula de código seguinte contém alguns exemplos de utilização da função.


```python
def busca(elem, itens):
  indice = 0
  num_itens = len(itens)
  resultado = None

  while indice < num_itens:

    if itens[indice] == elem:
      resultado = indice
      break
    indice = indice + 1
  return resultado

print(busca(2, [1, 2, 3]))
print(busca(3, [1, 2, 3]))
print(busca(4, [1, 2, 3]))
print(busca(1, [1, 2, 1]))

'''Nessa organização, o código começa definindo as variáveis num_itens, resultado e indice fora do loop while.
Em seguida, o loop while é definido corretamente com a verificação da condição indice < num_itens. Dentro do loop,
é verificado se itens[indice] == elem e, se a condição for verdadeira, o resultado é atualizado e o break é executado
para sair do loop. O indice é incrementado corretamente dentro do loop.
Por fim, o return resultado está corretamente alinhado com o escopo da função, retornando o valor atualizado de resultado após a conclusão do loop while.'''

```


## Problema 3

A célula abaixo contém uma função ``conta2`` que recebe um _string_ ``seq``, a função calcula o número de ocorrências em ``seq`` do padrão formado pelos dois primeiros caracteres de ``seq`` (incluindo sua ocorrência inicial). Por exemplo: ``conta2(arara)`` devolve 2. Caso ``seq`` tenha menos de 2 caracteres, a função devolve 0.

```python
#Resposta: 2,3,1,0

def conta2(seq):
  contador = 1
  indice = 2
  num_caracteres = len(seq)

  while indice < num_caracteres-1:
    if num_caracteres < 2:
      return 0
    else:
      if seq[indice] == seq[0] and seq[indice+1] == seq[1]:
        contador = contador + 1
      indice = indice + 1
  return contador


print(conta2("arara"))
print(conta2("ttgagcttgccaattaa"))
print(conta2("ga"))
print(conta2("a"))
```
