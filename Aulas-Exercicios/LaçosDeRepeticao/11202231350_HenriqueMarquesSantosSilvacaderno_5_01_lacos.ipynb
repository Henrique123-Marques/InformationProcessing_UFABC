## Exercício 1

Uma sequência de números $a_1, a_2, a_3, \ldots, a_n$ é considerada _crescente_ caso nenhum número da sequência seja menor que seu antecessor. Por exemplo: $1, 3, 3, 4, 6, 8, 8$ é uma sequência crescente; $2, 4, 6, 3, 8, 12$ não é uma sequência crescente.

Crie um programa python que:
1. Leia uma sequência de números digitados pelo usuário. A sequência encerra-se quando o usuário digitar o valor 0 (zero). O zero final _não é considerado parte da sequência_.
2. O seu programa deve imprimir se a sequência digitada é crescente ou não.

Use a célula abaixo para inserir sua resposta.


```python
def sequencia(numeros):
  lista = []
  while numeros != 0:
    numeros = int(input('Digite um numero: '))
    lista.append(numeros)
    if numeros == 0:
      lista.pop(-1)
      if all(lista[i] <= lista[i+1] for i in range(len(lista)-1)):
        return f'Programa encerrado, pois voce digitou 0, lista de numeros digitados: {lista}, está em ordem crescente'
      else:
        return f'Programa encerrado, pois voce digitou 0, lista de numeros digitados: {lista}, não está em ordem crescente'
print(sequencia(1))
```

## Exercício 4

Faça um programa que leia 10 números fornecidos pelo usuário e imprima a raiz quadrada de cada número. Caso o número digitado seja negativo, o seu programa deve imprimir uma mensagem de erro e continuar processamento os próximos números digitados pelo usuário.

```python
from math import sqrt

def calcular_raizes():
    lista_numeros = []
    for _ in range(10):
        numero = float(input('Digite um numero: '))
        if numero < 0:
            print('Valor negativo é invalido.')
        lista_numeros.append(numero)
    raizes = [round(sqrt(numero), 2) if numero >= 0 else "NaN" for numero in lista_numeros]
    return f'Raizes quadradas dos valores digitados: {raizes}'

print(calcular_raizes())
```

## Exercício 6

Crie uma função que calcula o _fatorial_ de um número inteiro positivo.


```python
def fatorial(aux):
  num = 1
  for i in range(1, aux+1):
    num *= i
  return num

fatorial(int(input('Digite um numero para saber seu fatorial: ')))
```

## Exercício 7

Uma sequência de números $a_1, a_2, a_3, \ldots, a_n$ contém uma _inversão_ quando um número não-negativo é sucedido por um número negativo. Por exemplo, na sequência a seguir, as inversões foram assinaladas em negrito: $1, 3, 0, \mathbf{-1}, 6, \mathbf{-3}, -1, 2, 8$.
Crie um programa python que:
1. Leia uma sequência de 8 números digitados pelo usuário.
2. O seu programa deve imprimir se a sequência contém alguma inversão ou não.

Use a célula abaixo para inserir sua resposta.

```python
lista = list()
def inversao_func():
  for i in range(8):
    lista.append(int(input('Digite um numero: ')))
  inversao = False

  for i in range(1, len(lista)):
    if lista[i-1] >= 0 and lista[i] < 0:
      inversao = True
      break

  if inversao:
    print(f'Tem inversao')
  else:
    print(f'Nao tem inversao')

inversao_func()
```

## Exercício 8

Faça um programa python que:
1. Leia uma sequência de números inteiros digitados pelo usuário, terminada com o número 0 (zero).
2. Imprima a somatória dos números _ímpares_ digitados pelo usuário.

Use a célula abaixo para inserir sua resposta.

```python
import math

impares = []
def soma_impar(numeros, somatoria):
  while True:
    numeros = int(input('Digite um numero: '))
    if numeros == 0:
      break
    elif numeros < 0:
      return f'Valor inválido, por ser negativo'
    elif numeros % 2 == 1:
      impares.append(numeros)
      somatoria = sum(impares)
  return f'Os valores impares digitados foram: {impares} e a soma deles é: {somatoria}'

soma_impar(0,0)
```

## Exercício 9


Crie um programa python que:
1. Leia um número inteiro $n$ digitado pelo usuário.
2. Em seguida, leia uma sequência de números inteiros digitados pelo usuário, terminada por 0 (zero).
3. Seu programa deve imprimir a somatória dos números da sequência **que sejam maiores que** $n$.

Por exemplo, para $n= 4$ e a sequência $2, 7, 3, 6, 9, 1, 4, 3, 0$, seu programa deve imprimir $22$ ($7 + 6 + 9$)

Use a célula abaixo para inserir sua resposta.

```python
import math

lista = list()
def soma_n(n, numeros, somatoria):
  n = int(input('Digite um valor n: '))
  while True:
    numeros = int(input('Digite um numero: '))
    if numeros == 0:
      break
    elif numeros < 0:
      print(f'Valor invalido, por ser negativo')
    elif numeros > n:
      lista.append(numeros)
      somatoria = sum(lista)
  return f'Os valores maiores que n digitados são: {lista} e a somatoria deles é: {somatoria}'

soma_n(0,0,0)
```
