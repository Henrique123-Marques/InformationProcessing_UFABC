## Exercício 1

Faça uma função que recebe uma lista `arg` como argumento e devolve uma nova lista contendo os elementos de `arg` em ordem inversa. Por exemplo, se a função receber a lista `[1, 2, 3]`, o resultado devolvido deve ser `[3, 2, 1]`. Na solução do exercício, você deve percorrer explicitamente a lista de entrada (com um laço), você não pode usar apenas o recurso de indexação de listas do Python.

```python
import math

def nova_lista(arg):
  lista = []
  for i in arg:
    lista.append(i)
    invertida = lista[::-1]
  return invertida

nova_lista([1,2,3])
```

## Exercício 2

Crie sua própria função `max` que recebe uma lista de valores numéricos e devolve o maior valor contido na lista. Repita para a função `min` que devolve o menor valor da lista. Na solução do exercício, você deve percorrer explicitamente a lista de entrada (com um laço), você não pode simplesmente usar funções já prontas do Python.Teste as duas funções. Reflita: o que deveria acontecer caso sua função receba uma lista vazia como argumento?

```python
def max(lista):
  maior_valor = None
  if len(lista) > 0:
    maior_valor = 0
    for i in range(1, len(lista)):
      if lista[i] > lista[maior_valor]:
        maior_valor = i
    return lista[maior_valor]
  elif len(lista) == []:
    return []

max([1,2,3,4,5])
```

```python
def min(lista2):
  menor = None
  if len(lista2) > 0:
    menor = 0
    for j in range(1, len(lista2)):
      if lista2[j] < lista2[menor]:
        menor = j
    return lista2[menor]
  elif len(lista2) == []:
    return []

min([98,43,1,21,11,32])
```

```python
def indice_max(lista):
  imax = None
  if len(lista) > 0:
    imax = 0
    for i in range(1, len(lista)):
      if lista[i] > lista[imax]:
        imax = i
  return imax


if indice_max([1,2,3]) != 2:
  print('Erro')
if indice_max([2,3,1]) != 1:
  print('Erro')
if indice_max([3,2,1]) != 0:
  print('Erro')
```

## Exercício 4

Crie uma função `lista_fatorial` que recebe um número natural $n$ e devolve uma lista contendo os valores de $1!$ até $n!$. Reflita: o que deveria acontecer caso sua função receba um número negativo?

_Dica_: use o fato que $n! = n \times (n-1)!$ para $n > 1$.

```python
def lista_fatorial():
  n = 1
  lista = []
  while n > 0:
    n = int(input('Digite um numero: '))
    if n < 0:
      return 0
    if n > 0:
      fatorial = 1
      for i in range(1, n+1):
        fatorial *= i
  lista.append(fatorial)
  return lista

lista_fatorial()
```

## Exercício 5

Crie uma função que recebe uma lista de valores numéricos e devolve a média dos valores contidos na lista. Na solução do exercício, você deve percorrer explicitamente a lista de entrada (com um laço), você não pode simplesmente usar funções já prontas do Python. Reflita: o que deveria acontecer caso sua função receba uma lista vazia?

```python
def media(lista):
  soma = 0
  for i in lista:
    soma += i
  return soma/len(lista)

media([1,2,3,4])
```

## Exercício 6

Crie uma função que recebe uma lista $\text{v}$ de valores numéricos e devolve um valor lógico (booleano) indicando se a lista é crescente ou não. Uma lista numérica $\text{v}$ é crescente caso $\text{v}[i] \leq \text{v}[j]$ para todo $i < j$.

```python
def lista_v(lista):
  for i in range(1, len(lista)):
    if lista[i] < lista[i-1]:
      return False
  return True

lista_v([1,2,3,4,5,8,7,9])
```

## Exercício 7

Considere a seguinte situação: nós escolhemos aleatoriamente um conjunto de 366 pessoas. Cada pessoa anota sua data de aniversário (mês e dia) em um pedaço de papel e deposita o papel em uma urna. Nós começamos a sortear os papéis da urna, sem reposição, anotando quais datas de aniversário já foram sorteadas. O processo termina quando a data sorteada for igual a alguma data sorteada anteriormente (isto é, o processo encerra quando ocorre a primeira repetição). Nesse momento, anotamos quantas pessoas foram sorteadas (incluindo a última, com a data de aniversário repetida).

Crie uma função `simula_aniversario` que simula a situação descrita acima e devolve o número pessoas sorteadas. Você pode usar uma lista de 365 posições para registrar quais datas já foram sorteadas. Em vez de sortear datas (como 31/janeiro ou 20/setembro), você pode sortear números de 0 a 364 (inclusive) representando as datas do ano (ou seja, zero representa 01/janeiro e 364 representa 31/dezembro). Na simulação, suponha que não existam anos bissextos (isto é, existem 365 datas diferentes) e que todas as datas são igualmente prováveis (isto é, não existem épocas do ano que concentrem mais aniversários do que outras).

Faça em seguida um programa que chama a sua função 200 vezes e calcula a média dos valores devolvidos pela função.

Essa simulação é relacionada ao chamado _problema do aniversário_ (ou _paradoxo do aniversário_). O problema pergunta qual o menor número de pessoas que precisamos juntar para que a probabilidade de encontrar duas pessoas que compartilham a data de aniversário seja 50%.

O trecho de código abaixo demonstra como podemos sortear um número entre 0 e 364, de maneira equiprovável.

```python
# pacote do python relacionado a sorteios aleatórios
import random

# cria uma lista contendo números entre 0 a 364 (inclusive)
datas = list(range(0, 365))

# sorteia uma amostra (um elemento) da lista
for i in range(200):
  n = random.sample(datas, 365)
print(f'{len(n)} datas foram sorteadas')


# execute esta célula algumas vezes para observar que os valores
# sorteados vão se alternando a cada vez
```

```python
def simula_aniversario():
  datas_repetidas = list()
  for data in datas:
    if datas.count(data) > 1 and data not in datas_repetidas:
      datas_repetidas.append(data)
  print(f'Datas repetidas: {datas_repetidas}') #Use return

simula_aniversario()
```
