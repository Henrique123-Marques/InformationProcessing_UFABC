
## Exercício 1

Faça um programa que leia uma sequência de **números inteiros**, **positivos** ou **negativos**, que é encerrada com o número **zero**. O programa deve somar os valores **negativos** e imprimir ao final o valor dessa somatória. O zero final não é considerado parte da sequência. Por exemplo, na sequência: (9, -5, 2, -9, 12, 10, 7,  3, -2, 0)  a resposta procurada é -16. Note que não é necessário armazenar todos os números.

**Entrada**: sequência de números inteiros de tamanho desconhecido, **terminada por zero**.

**Saída**: somatória dos números **negativos** digitados pelo usuário.

```python
import math

lista = list()

def somatoria_negativos(numeros, soma):
  while True:
    numeros = int(input('Digite um numero: '))
    if numeros < 0:
      lista.append(numeros)
      soma = sum(lista)
    if numeros == 0:
      break
  return soma


somatoria_negativos(0, 0)
```

Em um país *X*, a moeda nacional chama-se *conto*. O país aplica uma taxa de alfândega progressiva para bens declarados ao se entrar no país. Bens de valor até 500 (quinhentos) contos são taxados em 5%. Acima de 500 contos, o valor excedente (a quantidade superior a 500) é taxado em 8%. Por exemplo:

- um bem de 100 contos recebe uma taxa de 5 contos.
- um bem de 500 contos recebe uma taxa de 25 contos.
- um bem de 600 contos recebe uma taxa de 33 contos (25 + 8).

Faça um programa que recebe um número positivo do tipo _float_, representando o valor de um bem e em seguida imprime o **valor da taxa aplicada**. Por exemplo, para a entrada 600,00 seu programa deve imprimir 33,00 (não se preocupe com a quantidade de casas decimais ao imprimir).

**Entrada**: um número positivo do tipo float, representando um valor em contos.

**Saída**: o valor da taxa aplicada ao valor de entrada.

```python
def pais_x(n, taxa):
  n = float(input('Digite o valor do bem: '))
  if n > 0 and n <= 500:
    taxa = n - (n * 5/100)
  elif n > 500 and n < 600:
    taxa = n * 8/100
  elif n >= 600:
    taxa = n * 5.5/100
  return f'{taxa:.2f}'

'''600 = 100%
    33 = x

    x = 3300/600'''

pais_x(0,0)
```

## Exercício 6

Uma sequência de números $a_1, a_2, a_3, \ldots, a_n$ contém uma inversão quando um número não-negativo é sucedido por um número negativo. Por exemplo, na sequência a seguir, as inversões foram assinaladas em negrito: $1, 3, 0, \mathbf{-1}, 6, \mathbf{-3}, -1, 2, 8$.
Crie um programa python que:
1. Leia uma sequência de 8 números digitados pelo usuário.
2. O seu programa deve imprimir se a sequência contém alguma inversão ou não.

Para que a resposta seja considerada, seu programa deve ler os números e iterar sobre eles usando um laço de repetição.


```python
lista = list()
def inversao_func():
  for i in range(8):
    lista.append(int(input('Digite um numero: ')))
  inversao = False

  for i in range(1, len(lista)):
    if lista[i-1] >= 0 and lista[i] < 0:
      inversao = True
      break

  if inversao:
    print(f'Tem inversao')
  else:
    print(f'Nao tem inversao')

inversao_func()
```

Crie um programa python que:
1. Leia um número inteiro $n$ digitado pelo usuário.
2. Em seguida, leia uma sequência de números inteiros digitados pelo usuário, terminada por 0 (zero).
3. Seu programa deve imprimir a somatória dos números da sequência **que sejam estritamente maiores que** $n$.

Por exemplo, para $n= 4$ e a sequência $2, 7, 3, 6, 9, 1, 4, 3, 0$, seu programa deve imprimir $22$ (isto é, $7 + 6 + 9$)

```python
import math

lista = list()
def maiores(n, m, soma):
  n = int(input('Digite um numero: '))
  while True:
    m = int(input('Digite um numero para a lista: '))
    if m > n:
      lista.append(m)
      soma =  sum(lista)
    if m == 0:
      break

  return f'Soma dos numeros que são maiores que {n}: {soma}'

maiores(0,0,0)
```

## Exercício 10

Uma empresa produz peças com tamanho médio de  100mm, com desvio máximo aceitável de 5mm para mais ou para menos. Peças fora desse padrão são descartadas. Crie um programa Python que leia uma sequência de números finalizada por $-1$ (esse último valor não é considerado parte da sequência). Cada número representa o tamanho de uma peça dentro de um lote. Seu programa deve contar o número de peças total e o número de peças fora do padrão e imprimir a porcentagem de peças reprovadas.

```python
lista_pecas = list()
n_pecas_fora_padrao = list()
def empresa(n_pecas):
  while True:
    n_pecas = int(input('Digite o tamanho da peça: '))
    if n_pecas == -1:
      break
    lista_pecas.append(n_pecas)

  for i in lista_pecas:
    if i > 105 or i < 95:
      n_pecas_fora_padrao.append(i)
  porcentagem_reprovadas = len(n_pecas_fora_padrao)/len(lista_pecas) *100
  return f'Tamanho das peças: {lista_pecas}; Número de peças total: {len(lista_pecas)}; Número de peças fora do padrão: {len(n_pecas_fora_padrao)}; Porcentagem de peças reprovadas: {porcentagem_reprovadas:.2f}%'

empresa(0)
```

Faça um programa que leia uma sequência de números inteiros, que é encerrada com o número zero. O programa deve calcular e apresentar a soma de todos os valores informados que forem **menores que zero**, ou então imprimir uma mensagem quando não ocorreram valores menores que zero.

```python
import math


def soma_negativos(numeros):
  lista = list()
  negativos = list()
  while True:
    numeros = int(input('Digite um número: '))

    lista.append(numeros)

    if numeros < 0:
      negativos.append(numeros)

    if numeros == 0:
      lista.pop(-1)
      break

  return f'lista: {lista}, numeros negativos da lista: {negativos}, Soma dos numeros negativos: {sum(negativos)}'

soma_negativos(0)
```
