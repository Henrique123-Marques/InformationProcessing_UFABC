## Exercício 1

Definimos a *inversão* de uma sequência *S* de caracteres como sendo a sequência obtida ao percorrer *S* do fim para o começo (da direita para a esquerda). Por exemplo, a inversão de "cabeça" é "açebac". Um *palíndromo* é uma palavra ou sequência de palavras que é idêntica à sua própria inversão. São exemplos de palíndromos:
- ovo
- arara
- socorram me subi no onibus em marrocos

Crie uma função que recebe um string ``texto`` como argumento e decide se o string é um palíndromo ou não. Sua função deve devolver o valor booleano ``True`` caso ``texto`` seja um palídromo, e ``False`` caso contrário. Reflita sobre os seguintes pontos:
- Como a função deve se comportar quando receber um string vazio?
- Como a função deve se comportar quando receber um string composto por um único caractere?
- Como sua função deve se comportar com relação a caracteres maiúsculos/minúsculos?

Para criar a função, você pode optar por um dos caminhos abaixo:

1 - Construir a inversão de ``texto`` e comparar o resultado com ``texto``.
2 - Crie um laço que compara o primeiro carácter de ``texto`` com o último, depois o segundo carácter com o penúltimo e assim por diante, até o meio do _string_.

Em seguida, crie um programa que pede ao usuário que digite um texto e então imprima se o texto digitado é um palíndromo ou não.

````python
def inversao(texto = str(input('Digite um texto: '))):
  if texto == ' ':
    return f'O texto é invalido por ser vazio'
  for i in texto:
    palindromo = texto[::-1]
    if palindromo == texto:
      return True
    else:
      return False


inversao()
````

## Exercício 2

Em um programa corretor ortográfico, uma das regras diz que após um sinal de pontuação (ponto final, ponto de exclamação, ponto de interrogação) deve vir um espaço e a letra seguinte deve ser maiúscula. Crie uma função que recebe um string, representando um parágrafo de texto, e devolve um valor booleano sinalizando se o string contém uma violação dessa regra.


````python
def corretor(texto):
  pontuacoes = ['.', '!', '?']
  for indice, caractere in enumerate(texto):
    if caractere in pontuacoes:
      if indice + 1 < len(texto):
        if texto[indice + 1] != ' ' or not texto[indice + 2].isupper():
          return False
  return True



corretor('Brasil! argentina!')
````

## Exercício 3

A função ``split`` pode ser usada para dividir um _string_, devolvendo uma lista de suas partes (o _string_ original não é alterado). Por exemplo, o seguinte trecho:

```
frase = "Hoje é noite de lua cheia"
palavras = frase.split()
# palavras agora contém a lista
# ["Hoje", "é", "noite", "de", "lua", "cheia"]
```

Crie uma função ``inicial_caixa_alta`` que recebe um _string_ de texto e devolve uma versão do _string_ em que cada palavra começa com letra maiúscula. Por exemplo, ``inicial_caixa_alta("Choveu muito ontem)`` deve devolver "Choveu Muito Ontem".

Para implementar a função, você deve:
- Separar o conteúdo do _string_  de entrada em palavras.
- Transformar cada palavra para a formatação com primeira letra maiúscula.
- Concatenar as palavras transformadas para obter a resposta.

Em seguida, crie um programa que pede ao usuário que digite um texto e em seguida imprima o texto no formato com letra maiúscula na inicial de cada palavra.

````python
def inicial_caixa_alta(texto):
    # Divide o texto em palavras
    texto_separado = texto.split()
    # Inicializa a variável frase como uma string vazia
    frase = ''
    # Itera sobre cada palavra no texto separado
    for i in texto_separado:
        # Deixa a primeira letra da palavra maiúscula e adiciona à frase
        frase = frase + i.capitalize() + ' '
    # Retorna a frase resultante
    return f'Frase final: {frase}'

inicial_caixa_alta('Choveu muito hoje')
````

## Exercício 4

O _Wordle_ é um jogo online de adivinhação de palavras que se tornou popular alguns anos atrás, gerando vários clones (como o term.ooo ou o Letreco, em português). Suponha que queremos  criar uma versão do jogo em python, interagindo com o jogador via texto. O jogador precisa adivinhar uma palavra secreta de 5 letras, previamente determinada. A cada tentativa, o programa informa ao jogador o grau de similaridade entre sua tentativa e a palavra secreta.
- caso o jogador tenha adivinhado uma letra em sua posição correta, o programa deve mostrar um cifrão "\$" naquela posição.
- caso uma letra da palavra tentada pertença à palavra secreta, mas na posição errada, o programa deve mostrar uma exclamação "!" na posição dessa letra na palavra tentada.
- caso uma letra **não** pertença à palavra secreta, o programa deve mostrar o carácter sublinhado "_" na posição dessa letra na palavra tentada.
- Caso a palavra tentada contenha letras repetidas que apareçam uma ou mais vezes na palavra secreta, o programa deve mostrar uma quantidade de acertos (na mesma posição ou não) igual ao número de vezes que a letra aparece na palavra secreta.

Suponha que a palavra secreta seja "pedra", a lista abaixo mostra algumas tentativas e a respectiva resposta do programa.
- para a tentativa correta (pedra), o programa deve mostrar "\$\$\$\$\$".
- para a tentativa "podre", o programa deve mostrar "\$\_\$\$!".
- para a tentiva "tampa", o programa deve mostrar "\_\_\_!\$".
- para a tentativa "amado", o programa deve mostrar "!\_\_!\_".
- para a tentativa "conto", o programa deve mostrar "\_\_\_\_\_".

Crie uma função chamada ``resposta(tentativa, segredo)`` que recebe como _strings_ a tentativa do jogador e a palavra secreta e devolve a resposta que o programa deve mostrar ao jogador.


````python
def resposta(tentativa, segredo):
    resultado = ''
    if tentativa == segredo:
        for i in tentativa:
            resultado += '$'
        return resultado
    elif tentativa != segredo:
        for i in tentativa:
          resultado += '_'
        return resultado

tentativa = input('Digite uma palavra: ')
print(resposta(tentativa, 'ola'))
````
