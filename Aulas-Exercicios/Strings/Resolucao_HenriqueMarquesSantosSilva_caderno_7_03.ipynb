É um caso frequente que textos encontrados na internet empreguem repetição de letras como expressão de exagero ou outro efeito, como em "muuuito" ou "beeem". Crie uma função que recebe um _string_ e produz uma cópia do conteúdo, mas eliminando casos de repetição de *3* letras ou mais. Por exemplo, o string "muuuito" deve produzir "muito", mas o string "voo" deve produzir "voo".



```python
def eliminando(texto): #Inicio da funcao
    resposta = ''
    i = 0
    while i < len(texto): #Enquanto o indice for menor que o tamanho do texto
        count = 1 #Adiciona um contador
        while i + count < len(texto) and texto[i] == texto[i + count]: #Enquanto o caractere somado com o contador for menor que o tamanho do texto e o caractere no texto for igual ao caractere nesse posicao
            count += 1
        if count < 3:
          resposta += texto[i:i+count] #Adiciona os caracteres repetidos na variavel resposta
        else:
          resposta += texto[i] #Adiciona apenas a o caractere do texto apenas uma vez
        i += count
    return resposta

texto = 'voo'
resposta = eliminando(texto)
print(resposta)

```
