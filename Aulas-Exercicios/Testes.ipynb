Exemplo 2 - Remoção de espaços vazios em uma string

````python
def remover_espacos(texto):
  removido = ''
  for i in texto:
    if i == ' ':
      removido = str(removido) - i
  return removido

````

````python

texto = 'abc'
esperado = 'a b c'
resposta = remover_espacos(texto)
if resposta != esperado:
  print('Erro para a string: ' + str(texto) + ', seus espacos nao foram removidos. O esperado era: '+ esperado + '. Resposta final: ' + str(resposta))
````

# **Listas**

````python
#Retornando o valor do indice
lista1 = [3,7,11]
lista2 = lista1[1:2] #vai do indice 0 até 1, e nao pega o indice 2
lista1[1] = 2
print(lista2)
````

````python
#Copia de uma lista
lista1 = [3,7,11]
lista2 = lista1[0:3] #vai do indice 0 até 2, e nao pega o indice 3 (indice que nao existe)
lista1[1] = 2
print(lista2)
````

````python
#lista é cres ou decres
lista = [3,7,11]
crescente  = True
tamanho_lista = len(lista)
for i in range(tamanho_lista-1): #o tamanho_lista-1 evita de extrapolar o tamanho da lista
  if lista[i] > lista[i+1]:
    crescente = False
print(crescente)
````

````python
#lista é cres ou decres - outra forma
lista = [3,7,11]
crescente  = True
tamanho_lista = len(lista)
for i in range(1, tamanho_lista): #lembre de procurar o porque dessas operacoes
  atual = lista[i]
  antecessor = lista[i-1]
  if atual < antecessor:
    crescente = False
print(crescente)

'''Atenção: Tome cuidado '''
````
