### Exercício 5

Queremos escrever um programa que recebe uma data, representada por uma
sequência de três números equivalentes ao dia ($d$), mês ($m$) e ano ($a$).
Em seguida, o programa deve imprimir um número, correspondente ao dia
da semana dessa data (0 para domingo, 1, para segunda-feira, ..., 5 sexta-feira,
6 para sábado).

Vamos usar as seguintes formulas, adequadas para
o calendário gregoriano ($d_0$ é o dia da semana calculado, domingo
é equivalente a zero, segunda-feira é equivalente a um e assim por diante):

\begin{equation}
  a_0 = a - (14 - m) / 12
\end{equation}
\begin{equation}
  x = a_0 + a_0/4 - a_0/100 + a_0/400
\end{equation}
\begin{equation}
m_0 = m + 12 * ((14 - m) / 12) - 2
\end{equation}
\begin{equation}
d_0 = (d + x + (31*m_0)/ 12) \mod 7
\end{equation}

Escreva na célula abaixo o código para calcular o dia da semana, transcrevendo as equações para a linguagem Python. Precisamos lembrar que a operação de divisão inteira é representada por `//` e que a operação *mod* (resto da divisão) é representada em python pelo operador `%`.

````python
#Dias da semana: Domingo = 0, Segunda-feira = 1, Terça-feira = 2, Quarta-feira = 3, Quinta-feira = 4, Sexta-feira = 5, Sábado = 6

#Variáveis
d, m, a = 4, 8, 2030

#Resolução
a0 = a -(14 - m)//12
x = a0 + a0//4 - a0//100 + a0//400
m0 = m + 12*((14-m)//12)-2
d0 = (d + x + (31 * m0)//12)%7
print(f'Ano: {a0} e Mês: {m}')
print(f'Dia da semana númerico: {d0}')

if d0 == 0:
  print('O dia da semana é domingo')
elif d0 == 1:
  print('O dia da semana é segunda-feira')
elif d0 == 2:
  print('O dia da semana é terça-feira')
elif d0 == 3:
  print('O dia da semana é quarta-feira')
elif d0 == 4:
  print('O dia da semana é quinta-feira')
elif d0 == 5:
  print('O dia da semana é sexta-feira')
elif d0 == 6:
  print('O dia da semana é sábado')
````
