 # Exercício

 Crie um programa que implemente o jogo "Pedra, Papel e Tesoura"
(também conhecido como "joquempô"). A mecânica do jogo é a seguinte: um usuário humano joga contra o programa. O programa deve pedir ao usuário que escolha uma das 3 opções possíveis (pedra, papel ou tesoura). O programa também deve realizar sua própria opção, aleatoriamente. Em seguida, o programa avalia quem foi o vencedor, com base na seguinte regra:
- pedra perde para papel
- papel perde para tesoura
- tesoura perde para pedra.

Caso ambos jogadores escolham a mesma opção, há empate. O programa deve então imprimir a opção de cada jogador e dizer quem foi vencedor (ou se houve empate).

A célula de código abaixo mostra como gerar um número inteiro aleatório dentro de uma faixa de valores. A função `random.randint(min, max)` devolve um número inteiro no intervalo fechado $[min, max]$ (isto é, ambos os extremos estão incluídos).

```python
import random

# execute esta célula algumas vezes em sequência e
# você notará que o valor impresso varia entre
# execuções
opcao = int(input('Vamos jogar joquempô, escolha o número 1 para pedra, 2 para papel ou 3 para tesoura: '))
x = random.randint(1, 3)
print(f'Opção do jogador: {opcao}, Opção da máquina: {x}')

if opcao == 1 and x == 2:
  print('Você escolheu pedra e eu escolhi papel, você perdeu, tente novamente!')
elif opcao == 1 and x == 3:
  print('Você escolheu pedra e eu escolhi tesoura, você venceu!')
elif opcao == 2 and x == 1:
  print('Você escolheu papel e eu escolhi pedra, você venceu!')
elif opcao == 2 and x == 3:
  print('Você escolheu papel e eu escolhi tesoura, você perdeu, tente novamente!')
elif opcao == 3 and x == 1:
  print('Você escolheu tesoura e eu escolhi pedra, você perdeu, tente novamente!')
elif opcao == 3 and x == 2:
  print('Você escolheu tesoura e eu escolhi papel, você venceu!')
else:
  print('Opções iguais, empatamos!')
```
